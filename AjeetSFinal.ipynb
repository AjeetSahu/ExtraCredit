```scala
val sqlContext = new org.apache.spark.sql.SQLContext(sc)
```

```scala
import sqlContext.implicits._
```

```scala
case class Building (LocationCode:String,RegionCode:String, BldgAddress1:String, BldgAddress2:String, BldgCity:String, BldgCounty:String, BldgState:String,BldgZip:String, CongressionalDistrict:String, BldgStatus:String, PropertyType:String, BldgANSIUsable:String, TotalParkingSpaces:String, OwnedLeased:String, HistoricalType:String, HistoricalStatus:String, ABAAccessibilityFlag:String)

```

```scala
val BuildingNoPark = sc.textFile("swift://notebooks.spark/datagovbldgrexus.csv").map(_.split(",")).map(p => Building(p(0),p(1),p(2),p(3),p(4),p(5),p(6),p(7),p(8),p(9),p(10),p(11),p(12),p(13),p(15),p(16),p(17))).toDF()
```

```scala
BuildingNoPark.registerTempTable("BuildingNoPark")
```

```scala
val park = sqlContext.sql("SELECT BldgAddress1,BldgAddress2, BldgCity FROM BuildingNoPark where TotalParkingSpaces = '0' ")
```

```scala
park.map(t => "Building Address: " + t(0) + " "+ t(1) + ", City:"+t(2)).collect().foreach(println)
```
